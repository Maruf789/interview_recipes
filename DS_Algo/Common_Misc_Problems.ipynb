```python
""" right rotate a list/array """

def rightRotateManual(lst, n):  # Solution #1: Manual Rotation (O(N), O(N))
    n = n%len(lst)
    rotatedList = [] 
    for item in range(len(lst) - n, len(lst)):
        rotatedList.append(lst[item]) 
    for item in range(0, len(lst) - n):  
        rotatedList.append(lst[item]) 
    return rotatedList 

def rightRotate(lst,n):   # Solution #2: Pythonic Rotation (O(1), O(1))
    n = n%len(lst)
    return lst[-n:] + lst[:-n]

print(rightRotate([1, 2, 3, 4, 5], 3))
print(rightRotateManual([1,2,3,4,5], 3))
print(rightRotate([300, -1, 3, 0], 1))
print(rightRotateManual([300, -1, 3, 0], 1))
print(rightRotate([0, 0, 0, 2], 1))
print(rightRotateManual([0, 0, 0, 2], 1))
print(rightRotate(['13', 'a', 'Python'],1))
print(rightRotateManual(['13', 'a', 'Python'],1))
```

```python
"""
determine is two strings are one edit away from each other -- My brute force solution
"""
def is_one_away(s1, s2):
    if len(s1) == len(s2):
        return is_one_away_same_length(s1, s2)
    elif len(s1) > len(s2):
        return is_one_away_varied_length(s1, s2)
    elif len(s1) < len(s2):
        return is_one_away_varied_length(s2, s1)
    return False
    
def is_one_away_same_length(s1, s2):
    n1, n2 = len(s1), len(s2)
    if n1 == 0 and n2 == 0:
        return False
    change = 0
    for i in range(0, n1):
        if s1[i] != s2[i]:
            change += 1

    if change ==0 or change == 1:
        return True
    else: return False

def is_one_away_varied_length(s1, s2):
    n1, n2 = len(s1), len(s2)
    if abs(n1-n2) >= 2: return False
    change = 0
    i, j = 0, 0
    while i < n1 and j < n2:
        if s1[i] != s2[j]:
            change += 1
            i += 1
        else:
            i+= 1
            j += 1
            
    if change ==0 or change == 1:
            return True
    else: return False

# NOTE: The following input values will be used for testing your solution.
print(is_one_away("abcde", "abcd"))  # should return True
print(is_one_away("abde", "abcde"))  # should return True
print(is_one_away("a", "a"))  # should return True
print(is_one_away("abcdef", "abqdef"))  # should return True
print(is_one_away("abcdef", "abccef"))  # should return True
print(is_one_away("abcdef", "abcde"))  # should return True
print(is_one_away("aaa", "abc"))  # should return False
print(is_one_away("abcde", "abc"))  # should return False
print(is_one_away("abc", "abcde"))  # should return False
print(is_one_away("abc", "bcc"))  # should return False
print(is_one_away("", ""))  # should return False
print(is_one_away("abc", ""))  # should return False
```

```python
"""
===== Populate Fibonacci Series: Different implementations =====
"""

def fibr(n):  #recursive (slowest)
    if n == 0 or n == 1:
        return n
    else:
        return fibr(n - 1) + fibr(n - 2)
    
def fibrm(n):   #recursive w/memoization (less slower)
    memo = {0:0, 1:1}
    if n not in memo:
        memo[n] = fibrm(n-1) + fibrm(n-2)
    return memo[n]

def fibi(n):  #iterative (fast)
    old, new = 0, 1
    if n == 0:
        return 0
    for i in range(n-1):
        old, new = new, old + new
    return new

def fib(n): #iterative w/memoization (fast)
    memo = [0,1]
    for i in range(2,n+1):
        memo.append(memo[i-1] + memo[i-2])
    print("Entire sequence: ", memo)
    return memo[n]

memo = {1:1, 2:1}
def fib_memoization(n):
    if n <= 2:
        return 1
    if n in memo:
        return memo[n]
    else:
        memo[n] = fib(n-1) + fib(n-2)
    return memo[n]
    
# Test cases
print(fibr(9))
print(fibi(9))
print(fib_memoization(9))
print(fib(9))
print(fibrm(9))
```

```python
"""Find palindrome: time: O(n), space O(n)"""

def find_palindrome(s):
    if not s:
        return False
    return all(s[i] == s[~i] for i in range(0, len(s)//2))

print(find_palindrome("tusut"))
print(find_palindrome("marufurag"))
```

```python
"""find anagrams approach 1 (sort and compare) - time O(n(log(n))), space O(1)"""

def anagramSolution1(s1,s2):
    if len(s1) > len(s2):
        return False
    
    alist1 = list(s1)
    alist2 = list(s2)

    alist1.sort()
    alist2.sort()

    pos = 0
    matches = True

    while pos < len(s1) and matches:
        if alist1[pos]==alist2[pos]:
            pos = pos + 1
        else:
            matches = False

    return matches

print(anagramSolution1('abcde','edcba'))
print(anagramSolution1('abcdek','edcba'))
```

```python
"""find anagrams approach 2 (Count and compare) - time O(n), space O(n + m)"""

def anagramSolution2(s1,s2):
    
    c1 = [0]*26
    c2 = [0]*26

    for i in range(len(s1)):
        pos = ord(s1[i])-ord('a')
        c1[pos] += 1

    for i in range(len(s2)):
        pos = ord(s2[i])-ord('a')
        c2[pos] += 1

    j = 0
    valid = True
    while j<26 and valid:
        if c1[j]==c2[j]:
            j = j + 1
        else:
            valid = False

    return valid

print(anagramSolution2('apple','pleap'))
print(anagramSolution1('abcdek','edcba'))
```

```python
"""Converting an (decimal) Integer to a String in Any Base using Stack (iterative)""" O(n), O(n)

from collections import deque
stack = deque()

def toStr(n,base): 
    convertString = "0123456789ABCDEF"
    while n > 0:
        if n < base:
            stack.append(convertString[n])
        else:
            stack.append(convertString[n % base])
        n = n // base
    
    result = ""
    while len(stack) > 0:
        result += str(stack.pop())
        
    return result

print(toStr(11,2))
#print(toStr(1435,16))
```

```python
"""Version 1: Minimum coins needed for change in vending machine (recursive w/o memoization)"""

def recMC(coinValueList,change):
    minCoins = change
    if change in coinValueList:
        return 1
    else:
        for i in [c for c in coinValueList if c <= change]:
            numCoins = 1 + recMC(coinValueList,change-i)
            if numCoins < minCoins:
                minCoins = numCoins
    return minCoins

print(recMC([1,5,10,25],63))
```

```python
""" Version 2: Minimum coins needed for change in vending machine (recursive + memoization) """

def recursiceDC(coinValueList,change,knownResults):
    minCoins = change
    if change in coinValueList:
        knownResults[change] = 1
        return 1
    elif knownResults[change] > 0:
        return knownResults[change]
    else:
        for i in [c for c in coinValueList if c <= change]:
            numCoins = 1 + recursiceDC(coinValueList, change-i,
                              knownResults)
            if numCoins < minCoins:
                minCoins = numCoins
                knownResults[change] = minCoins
                
    return minCoins

knownResults = [0] * 64
print(recursiceDC([1,5,10,25],63, knownResults))
print(knownResults)
```

```python
# Python program for Boggle game - Trie based

class Boggle:
    """
    Boggle (Find all possible words in a board of characters)
    DFS-based: https://www.geeksforgeeks.org/boggle-find-possible-words-board-characters/
    Trie-based: https://www.geeksforgeeks.org/boggle-set-2-using-trie/
    """
    
    def __init__(self, dictionary, grid):
        self.dictionary = set(dictionary) # you could use a Trie instead
        self.boggle = grid
        self.grid_row = len(grid)
        self.grid_col = len(grid[0])
            
    def isWord(self, currentWord):
        """
        A given function to check if a given currentWord is present in 
        dictionary. The implementation is naive and costs O(N). 
        Alternatively, you could use a Trie for faster search.
        """
        return True if currentWord in self.dictionary else False
        
    def findWords(self):
        """Prints all words present in dictionary. """
        
        result = set()
        
        # Mark all characters as not visited 
        visited = [[False for col in range(self.grid_col)] for row in range(self.grid_row)] 

        # Initialize current currentWord 
        currentWord = ""

        # Consider every character and look for all words 
        # starting with this character 
        for i in range(0, self.grid_row):
            for j in range(0, self.grid_col):
                self.findWordsUtil(result, visited, i, j, currentWord)
        
        return result
        

    def findWordsUtil(self, result, visited, i, j, currentWord):
        """A recursive function to print all words present on boggle."""
        
        # Mark current cell as visited and append current characters to currentWord 
        visited[i][j] = True
        currentWord += self.boggle[i][j]

        # If str is present in dictionary, then print it 
        if self.isWord(currentWord):
            result.add(currentWord)

        # Traverse 8 adjacent cells of boggle[i][j] 
        for row in range(i-1, self.grid_row):
            for col in range(j-1, self.grid_col):
                if (row >= 0 and col >= 0 and visited[row][col] == False):
                    self.findWordsUtil(result, visited, row, col, currentWord)

        # Erase current character from currentWord and mark visited 
        # of current cell as False 
        currentWord = "" + currentWord[len(currentWord)-1]
        visited[i][j] = False

# Driver program to test above function 
def main():
    # Let the given dictionary be following 
    dictionary = [ "GEEKS", "FOR", "QUIZ", "GUQ", "EE" ]
    grid = [[ 'G', 'I', 'Z' ],
            [ 'U', 'E', 'K' ], 
            [ 'Q', 'S', 'E' ]]
    
    boggle = Boggle(dictionary, grid)

    print("Following words of dictionary are present")
    print("List of words found in the boggle: ", boggle.findWords())

main()
```
