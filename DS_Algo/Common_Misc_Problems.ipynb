```python
"""determine the non-repeating character in a string"""
def non_repeating(given_string):
    
    m = {}
    for i in given_string:
        if i not in m.keys():
            m[i] = 1
        else:
            m[i] += 1
    
    for i in given_string:
        if m[i] == 1:
            return i
            
    return None

# NOTE: The following input values will be used for testing your solution.
print(non_repeating("abcab"))   # should return 'c'
print(non_repeating("abab"))    # should return None
print(non_repeating("aabbbc"))  # should return 'c'
print(non_repeating("aabbdbc")) # should return 'd'
```

```python
""" return the intersection array (common elements between two arrays)"""
def common_elements(list1, list2):
    result = []
    
    p1 =0;
    p2 =0;
    
    while p1 < len(list1) and p2 < len(list2):
        if list1[p1] == list2[p2]:
            result.append(list1[p1])
            p1 += 1
            p2 += 1
        elif list1[p1] > list2[p2]:
            p2 += 1
        else:
            p1 += 1
    
    return result


# NOTE: The following input values will be used for testing your solution.
list_a1 = [1, 3, 4, 6, 7, 9]
list_a2 = [1, 2, 4, 5, 9, 10]
print(common_elements(list_a1, list_a2)) #should return [1, 4, 9] (a list).

list_b1 = [1, 2, 9, 10, 11, 12]
list_b2 = [0, 1, 2, 3, 4, 5, 8, 9, 10, 12, 14, 15]
print(common_elements(list_b1, list_b2)) #should return [1, 2, 9, 10, 12] (a list).

list_c1 = [0, 1, 2, 3, 4, 5]
list_c2 = [6, 7, 8, 9, 10, 11]
print(common_elements(list_c1, list_c2)) #should return [] (an empty list).
```

```python
"""determine if two arrays are rotations of each other"""
def is_rotation(list1, list2):
    if len(list1) != len(list2):
        return False
    
    key_list2 = -1
    key = list1[0]
    
    for i in range(0, len(list2)):
        if key == list2[i]:
            key_list2 = i
            break
    
    for i in range(0, len(list1)):
        j = (key_list2 + i) % len(list1)
        if list1[i] != list2[j]:
            return False
    
    return True

# NOTE: The following input values will be used for testing your solution.
list1 = [1, 2, 3, 4, 5, 6, 7]
list2a = [4, 5, 6, 7, 8, 1, 2, 3]
print(is_rotation(list1, list2a)) #should return False.
list2b = [4, 5, 6, 7, 1, 2, 3]
print(is_rotation(list1, list2b)) #should return True.
list2c = [4, 5, 6, 9, 1, 2, 3]
print(is_rotation(list1, list2c)) #should return False.
list2d = [4, 6, 5, 7, 1, 2, 3]
print(is_rotation(list1, list2d)) #should return False.
list2e = [4, 5, 6, 7, 0, 2, 3]
print(is_rotation(list1, list2e)) #should return False.
list2f = [1, 2, 3, 4, 5, 6, 7]
print(is_rotation(list1, list2f)) #should return True.
```

```python
"""
determine is two strings are one edit away from each other -- My brute force solution
"""
def is_one_away(s1, s2):
    if len(s1) == len(s2):
        return is_one_away_same_length(s1, s2)
    elif len(s1) > len(s2):
        return is_one_away_varied_length(s1, s2)
    elif len(s1) < len(s2):
        return is_one_away_varied_length(s2, s1)
    return False
    
def is_one_away_same_length(s1, s2):
    n1, n2 = len(s1), len(s2)
    if n1 == 0 and n2 == 0:
        return False
    change = 0
    for i in range(0, n1):
        if s1[i] != s2[i]:
            change += 1

    if change ==0 or change == 1:
        return True
    else: return False

def is_one_away_varied_length(s1, s2):
    n1, n2 = len(s1), len(s2)
    if abs(n1-n2) >= 2: return False
    change = 0
    i, j = 0, 0
    while i < n1 and j < n2:
        if s1[i] != s2[j]:
            change += 1
            i += 1
        else:
            i+= 1
            j += 1
            
    if change ==0 or change == 1:
            return True
    else: return False

# NOTE: The following input values will be used for testing your solution.
print(is_one_away("abcde", "abcd"))  # should return True
print(is_one_away("abde", "abcde"))  # should return True
print(is_one_away("a", "a"))  # should return True
print(is_one_away("abcdef", "abqdef"))  # should return True
print(is_one_away("abcdef", "abccef"))  # should return True
print(is_one_away("abcdef", "abcde"))  # should return True
print(is_one_away("aaa", "abc"))  # should return False
print(is_one_away("abcde", "abc"))  # should return False
print(is_one_away("abc", "abcde"))  # should return False
print(is_one_away("abc", "bcc"))  # should return False
print(is_one_away("", ""))  # should return False
print(is_one_away("abc", ""))  # should return False
```

```python
"""
===== Populate Fibonacci Series: Different implementations =====
"""

def fibr(n):  #recursive (slowest)
    if n == 0 or n == 1:
        return n
    else:
        return fibr(n - 1) + fibr(n - 2)
    
def fibrm(n):   #recursive w/memoization (less slower)
    memo = {0:0, 1:1}
    if n not in memo:
        memo[n] = fibrm(n-1) + fibrm(n-2)
    return memo[n]

def fibi(n):  #iterative (fast)
    old, new = 0, 1
    if n == 0:
        return 0
    for i in range(n-1):
        old, new = new, old + new
    return new

def fib(n): #iterative w/memoization (fast)
    memo = [0,1]
    for i in range(2,n+1):
        memo.append(memo[i-1] + memo[i-2])
    print("Entire sequence: ", memo)
    return memo[n]

memo = {1:1, 2:1}
def fib_memoization(n):
    if n <= 2:
        return 1
    if n in memo:
        return memo[n]
    else:
        memo[n] = fib(n-1) + fib(n-2)
    return memo[n]
    
# Test cases
print(fibr(9))
print(fibi(9))
print(fib_memoization(9))
print(fib(9))
print(fibrm(9))
```

```python
"""Find palindrome: time: O(n), space O(n)"""

def find_palindrome(s):
    if not s:
        return False
    return all(s[i] == s[~i] for i in range(0, len(s)))

print(find_palindrome("tusut"))
print(find_palindrome("marufurag"))
```

```python
"""find anagrams approach 1 (sort and compare) - time O(n(log(n))), space O(1)"""

def anagramSolution1(s1,s2):
    if len(s1) > len(s2):
        return False
    
    alist1 = list(s1)
    alist2 = list(s2)

    alist1.sort()
    alist2.sort()

    pos = 0
    matches = True

    while pos < len(s1) and matches:
        if alist1[pos]==alist2[pos]:
            pos = pos + 1
        else:
            matches = False

    return matches

print(anagramSolution1('abcde','edcba'))
print(anagramSolution1('abcdek','edcba'))
```

```python
"""find anagrams approach 2 (Count and compare) - time O(n), space O(n + m)"""

def anagramSolution2(s1,s2):
    
    c1 = [0]*26
    c2 = [0]*26

    for i in range(len(s1)):
        pos = ord(s1[i])-ord('a')
        c1[pos] += 1

    for i in range(len(s2)):
        pos = ord(s2[i])-ord('a')
        c2[pos] += 1

    j = 0
    valid = True
    while j<26 and valid:
        if c1[j]==c2[j]:
            j = j + 1
        else:
            valid = False

    return valid

print(anagramSolution2('apple','pleap'))
print(anagramSolution1('abcdek','edcba'))
```

```python
"""Converting an (decimal) Integer to a String in Any Base using Stack (iterative)""" O(n), O(n)

from collections import deque
stack = deque()

def toStr(n,base): 
    convertString = "0123456789ABCDEF"
    while n > 0:
        if n < base:
            stack.append(convertString[n])
        else:
            stack.append(convertString[n % base])
        n = n // base
    
    result = ""
    while len(stack) > 0:
        result += str(stack.pop())
        
    return result

print(toStr(11,2))
#print(toStr(1435,16))
```

```python
"""Version 1: Minimum coins needed for change in vending machine (recursive w/o memoization)"""

def recMC(coinValueList,change):
    minCoins = change
    if change in coinValueList:
        return 1
    else:
        for i in [c for c in coinValueList if c <= change]:
            numCoins = 1 + recMC(coinValueList,change-i)
            if numCoins < minCoins:
                minCoins = numCoins
    return minCoins

print(recMC([1,5,10,25],63))
```

```python
""" Version 2: Minimum coins needed for change in vending machine (recursive + memoization) """

def recursiceDC(coinValueList,change,knownResults):
    minCoins = change
    if change in coinValueList:
        knownResults[change] = 1
        return 1
    elif knownResults[change] > 0:
        return knownResults[change]
    else:
        for i in [c for c in coinValueList if c <= change]:
            numCoins = 1 + recursiceDC(coinValueList, change-i,
                              knownResults)
            if numCoins < minCoins:
                minCoins = numCoins
                knownResults[change] = minCoins
                
    return minCoins

knownResults = [0] * 64
print(recursiceDC([1,5,10,25],63, knownResults))
print(knownResults)
```

```python
"""??Version 3: Minimum coins needed for change in vending machine (dynamic prog)"""

def dpMakeChange(coinValueList,amount,minCoins,coinsUsed):
    for cents in range(amount+1):
        coinCount = cents
        newCoin = 1
        for j in [c for c in coinValueList if c <= cents]:
            if minCoins[cents-j] + 1 < coinCount:
                coinCount = minCoins[cents-j]+1
                newCoin = j
        minCoins[cents] = coinCount
        coinsUsed[cents] = newCoin
    
    return minCoins[amount]

def printCoins(coinsUsed,amount):
    coin = amount
    while coin > 0:
        thisCoin = coinsUsed[coin]
        print(thisCoin)
        coin = coin - thisCoin

def main():
    amount = 63
    coinValueList = [1,5,10,25]
    coinsUsed = [0]*(amount+1)
    minCoins = [0]*(amount+1)

    print("Making change for", amount, "requires --")
    print(dpMakeChange(coinValueList, amount, minCoins, coinsUsed), "coins")
    print("They are:")
    printCoins(coinsUsed, amount)
    print("The coin used list is as follows:")
    print(coinsUsed)
    print("The coin count list is as follows:")
    print(minCoins)

main()
```
