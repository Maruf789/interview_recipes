```python
def permutation(pattern):
  l = []
  if len(pattern) == 0:
    return []
  if len(pattern) == 1:
    return [pattern]
  for i in range(len(list(pattern))):
    m = pattern[i]
    rest = pattern[:i]+ pattern[i+1:]
    for p in permutation(rest):
      l.append(m + p)
  return l

def main():
  print('Permutation exist: ' + str(permutation("oidbcaf")))
  print('Permutation exist: ' + str(permutation("odicf")))
  print('Permutation exist: ' + str(permutation("bcdxabcdy")))
  print('Permutation exist: ' + str(permutation("aaacb")))


main()
```

```python
def permutation(arr):
  l = []
  if len(arr) == 0:
    return []
  if len(arr) == 1:
    return [arr]
  for i in range(len(arr)):
    m = arr[i]
    rest = arr[:i] + arr[i+1:]
    for p in permutation(rest):
      new_permutation = [''.join([m] + p)]
      l.append(new_permutation)
  return l


def main():
  print(permutation(["cat", "fox", "dog"]))

main()
```
