```python
"""Count the number of valleys and mountains

Example: "UDDDUDUU" can be represented as:
_/\      _
   \    /
    \/\/  There are 1 valley and 1 mountain in the above example, so answer is 1,1
"""
def countingValleysMountains(n, s):
    num_of_valleys = 0           # number of valleys
    num_of_mountains = 0         # number of mountains
    curr_level = 0               # current level
    for character in s:
        if(character == 'U'): curr_level += 1
        if(character == 'D'): curr_level -= 1
        if(curr_level == 0 and character == 'U'): # if we just came UP from valley to sea level
            num_of_valleys += 1
        if(curr_level == 0 and character == 'D'): # if we just came DOWN from mountain to sea level
            num_of_mountains += 1
    
    print("input steps: '" +  s + "'")
    print("\tnum_of_valleys: ", num_of_valleys)
    print("\tnum_of_mountains: ", num_of_mountains)
    return num_of_valleys, num_of_mountains

"""
_/\      _
   \    /
    \/\/
"""
print(countingValleysMountains(8, 'UDDDUDUU'))                    # expected 1,1
"""
_          /\
 \  /\    /
  \/  \/\/
         
"""
print(countingValleysMountains(12, 'DDUUDDUDUUUD'))               # expected 2,1

"""
          /\
   /\    /  \
  /  \  /    \
_/    \/      \
               \  /\      /
                \/  \    /
                     \  /
                      \/
"""
print(countingValleysMountains(12, 'UUUDDDUUUUDDDDDDUUDDDDUUUU')) # expected 2,2
```

```python
"""find all permutations of s within b"""
def find_all_permutations(s,b):
    from itertools import permutations 
    result = 0
    s_permutations = [''.join(list(each_permutation)) for each_permutation in list(permutations(s))]       
    for permutation in s_permutations:
            if permutation in b:
                result += 1   
    return result

print(find_all_permutations('xacxzaa', 'fxaazxacaaxzoecazxaxaz'))
```

```python
"""Hash Tables: Ransom Notes
Given two dictionaries, tell one of them is a subset of the other.
"""

def checkMagazine(magazine, note):
    result = False
    magazine_dict = {}
    for word in magazine:
        if word not in magazine_dict: magazine_dict[word] = 1
        else: magazine_dict[word] += 1
    note_dict = {}
    for word in note:
        if word not in note_dict: note_dict[word] = 1
        else: note_dict[word] += 1
    for k,v in note_dict.items():
        if k in magazine_dict:
            if v <= magazine_dict[k]: result = True
            else: 
                result = False
                break
        else:
            result = False
            break
    if result: print("Yes")
    else: print("No")
        
checkMagazine("give me one grand today night".split(" "), "give one grand today".split(" "))
checkMagazine("ive got a lovely bunch of coconuts".split(" "), "ive got some coconuts".split(" "))
checkMagazine("two times three is not four".split(" "), "two times two is four".split(" "))
```
