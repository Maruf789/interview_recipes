```python
"""Jumping Clouds"""
def jumpingOnClouds(c):  #hackerrank solution
    n = len(c)
    print("\ninput array: ", c)
    ans = 0
    i = 0
    while i < n - 1:
        if i + 2 >= n or c[i + 2] == 1:   # Not possible to make a jump of size 2
            print(i, end=" ")
            i = i + 1
            ans = ans + 1
        else:
            print(i, end=" ")
            i = i + 2
            ans = ans + 1
            
    print("\nresult: ", end="")
    return ans

def crossingOnCloudsPossible(c):  #my optimization
    print("\ninput array: ", c)
    shortest_path = 0
    possible = False
    n = len(c)
    
    if n == 0: # not needed, just for the corner case
        print("shortest_path, possible_to_reach_end: ", end="") 
        return shortest_path, possible
    if n == 1:   # not needed, just for the corner cases
        print("shortest_path, possible_to_reach_end: ", end="")
        if c[0] == 0: 
            possible = True
            return shortest_path, possible
        else: 
            return shortest_path, possible
    
    i = 0
    print("path to destination: ", end=" ")
    while i < n - 1:
        if i + 2 >= n or c[i + 2] == 1: # reached the end OR Not possible to make a jump of size 2
            if c[i+1] == 0:             # possible to make a jump of size 1
                print(i, end=" ")
                i = i + 1
                possible = True
                shortest_path += 1
            else:                       # not possible to make jump of both size 1 and 2
                print(i, end=" ")
                possible = False
                break
        else:
            if c[i+2] == 0:             # possible to make a jump of size 2
                print(i, end=" ")
                i = i + 2
                possible = True
                shortest_path += 1
            else:                       # not possible to make a jump of size 2
                print(i, end=" ")
                i = i + 2
                possible = False
            
    print("\nshortest_path, possible_to_reach_end: ", end="")
    return shortest_path, possible
    
print(crossingOnCloudsPossible([]))
print(crossingOnCloudsPossible([0]))
print(crossingOnCloudsPossible([1]))
print(crossingOnCloudsPossible([0,0]))
print(crossingOnCloudsPossible([0,1,1]))
print(crossingOnCloudsPossible([0,0,0,0,0]))
print(crossingOnCloudsPossible([0,0,1,0,0,1,0]))
print(crossingOnCloudsPossible([0,0,0,1,0,0]))
print(crossingOnCloudsPossible([0,1,1,1,1]))
print(crossingOnCloudsPossible([0,1,1,0,0]))
```

```python
def repeatedString(s, n):
    print("input string: " + s + ", n: " + str(n))
    count = 0
    lenOfInputString = len(s)
    if lenOfInputString > 1:
        for letter in s:
            if letter == 'a':
                count += 1
        print("\tcount: ", count)
    else:
        if s == 'a' : return n
        else: return 0
    res = count
    lenOfRepeatedString = lenOfInputString
    while lenOfRepeatedString < n:
        lenOfRepeatedString = lenOfRepeatedString + lenOfInputString
        res = res + count
    print("\tlenOfRepeatedString: ", lenOfRepeatedString )
    print("\tres: ", res)
    if lenOfRepeatedString == n: return res
    if lenOfRepeatedString > n:   #we went one over (incremented once extra-- gottta remove it)
        lenOfExtraLettersExceedingN = lenOfRepeatedString - n
        res = res - count   #removing the extra iteration from the result first 
                            # and then count from the leftovers and to the result
        additional_string = s[:len(s)- lenOfExtraLettersExceedingN]
        print("\tadditional_string_to_make_up_n:", additional_string)
        number_of_a_in_additional_string = 0
        for letter in additional_string:
            if letter == 'a':
                number_of_a_in_additional_string += 1
        res = res + number_of_a_in_additional_string
        return res
    
print("result:", repeatedString('abaabaabaa', 10))
print("result:", repeatedString('aba', 10))
print("result:", repeatedString('a', 1000000000000))
print("result:", repeatedString('b', 1000000000000))
print("result:", repeatedString('akjbadalkfb', 1000))
```
